```python
from linkedin_api import Linkedin
from googlesearch import search
import pandas as pd

#Authenticate using any Linkedin account credentials

#Enter username of LinkedIn account
username = "to.adityasurana@gmail.com"
#Enter password of linkedIn account
password = "***"
api = Linkedin(username, password)

#Reading 'LinkedIn test.xlsx' file - name should be same and case sensitive
df = pd.read_excel("LinkedIn test.xlsx")
n = df.shape[1]
company = df['CompanyName']  #Column name of 'Company' should be same and case sensitive
state = df['State'] #Column name of 'State' should be same and case sensitive

for i in range(1, n-3+1):
    links = ['']*len(company)
    for j in range(len(company)):
        title = df['Title'+str(i)] #Column name of 'Title1, Title2, ... Title15' should be same and case sensitive
        companyWithState = company[j] + ',' + ' ' + state[j]
        print(companyWithState, title[j])
        
        heading = 'LinksWithTitle' + str(i)
        
        try:
            result = api.search_people(keyword_title = title[j], keyword_company = companyWithState)
        except:
            result = []
        url = []
        for res in result:
            url.append('https://www.linkedin.com/in/'+res['public_id']+'/')
            print('got url: ', 'https://www.linkedin.com/in/'+res['public_id']+'/')
        if len(url) > 1:
            links[j] = url
            df[heading] = links
            try:
                df.to_excel('LinkedIn-output-withTitle%s.xlsx' %str(i), index=False)
            except:
                pass
            
        elif len(url) == 1:
            links[j] = url[0]
            df[heading] = links
            try:
                df.to_excel('LinkedIn-output-withTitle%s.xlsx' %str(i), index=False)
            except:
                pass
        else:
            query = companyWithState + " " + title[j] + " " + "linkedin"
            try:
                print('google query -> ', query)
                linkArray = []
                for link in search(query, tld="co.in", num=3, stop=3, pause=10):
                    checkLink = link.split('/')
                    if checkLink[-2] == 'in':
                        print("google query result -> ", link)
                        linkArray.append(link)
                if len(linkArray) > 1:
                    links[j] = linkArray
                    df[heading] = links
                    try:
                        df.to_excel('LinkedIn-output-withTitle%s.xlsx' %str(i), index=False)
                    except:
                        pass
                elif len(linkArray) == 1:
                    links[j] = linkArray[0]
                    df[heading] = links
                    try:
                        df.to_excel('LinkedIn-output-withTitle%s.xlsx' %str(i), index=False)
                    except:
                        pass
                else:
                    pass
            except:
                pass
df.to_excel('LinkedIn-output.xlsx', index=False)
print("JOB COMPLETED, please find data in LinkedIn-output.xlsx")
```
